### Regresion lineal simple

#### Modelo de regresion lineal simple para 
##### x=Inversion en publicidad en TV 
###### y=total de ventas 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
datos=pd.read_csv('Advertising.csv')
```

```python
datos
```

```python
datos.head()
```

```python
datos=datos.drop('Unnamed: 0', axis=1)
```

```python
datos.info()
```

```python
datos.describe()
```

```python
sns.pairplot(datos)
```

```python
sns.displot(datos['Sales'], color='green')
```

```python
datos.corr()
```

```python
sns.heatmap(datos.corr(), annot=True)
```

```python
X=datos[['TV']]
```

```python
X
```

```python
y=datos['Sales']
```

```python
y
```

```python
datos.plot(kind='scatter', x='TV', y='Sales')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo=LinearRegression()
```

```python
modelo


```

```python
modelo.fit(X,y)
```

```python
modelo.intercept_
```

```python
print(modelo.intercept_)
```

```python
print(modelo.coef_)
```

$$y=7.03259+0.04753X$$

```python
prediccion=modelo.predict(X)
```

```python
prediccion
```

```python
plt.scatter(y, prediccion)
```

```python
sns.regplot(x=y, y=prediccion, data=datos)
```

```python
from sklearn import metrics
```

```python
MSE=metrics.mean_squared_error(y,prediccion)
```

```python
MSE
```

```python
RSS=MSE*200
```

```python
print(RSS)
```
