```python
from sklearn.linear_model import LinearRegression
```

```python
modelo=LinearRegression()
```

```python
modelo
```

```python
modelo.fit(datos[['Radio']], datos['Sales'])
```

```python
modelo.intercept_
```

$$B0=9.311638095158283$$

```python
modelo.coef_
```

$$B1=0.20249578$$

$$Y=B0+B1X$$

$$Y=9.311638095158283+0.20249578X$$

```python
prediccion=modelo.predict(datos[['Radio']])
```
